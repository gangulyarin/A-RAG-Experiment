```python
docs = [
    "Paris is the capital of France.",
    "France is in Europe.",
    "Berlin is the capital of Germany.",
    "The Eiffel Tower is in Paris."
]
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(docs)
index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(np.array(embeddings))
```

```python
def keyword_search(query):
    return [d for d in docs if query.lower() in d.lower()]

def semantic_search(query, k=2):
    q_emb = model.encode([query])
    D, I = index.search(np.array(q_emb), k)
    return [docs[i] for i in I[0]]

def read_chunk(chunk):
    return chunk  # trivial here
```

```python
from llama_cpp import Llama

llm = Llama(
    model_path="/home/arindam/TinyLlama/tinyllama-1.1b-chat-v1.0.Q3_K_M.gguf",  # download TinyLlama GGUF
    n_ctx=2048
)

def llm_decide(prompt):
    return llm(prompt, max_tokens=200)["choices"][0]["text"]
```

```python
def agent_llm(question):
    context = ""
    
    for step in range(3):
        prompt = f"""
You are a retrieval agent.

Question: {question}
Context so far: {context}

Choose one action:
1. keyword_search
2. semantic_search
3. answer

Respond with action and query.
"""
        decision = llm_decide(prompt)
        print("Decision:", decision)
        
        if "keyword" in decision:
            results = keyword_search(question)
        elif "semantic" in decision:
            results = semantic_search(question)
        else:
            return context
        
        context += " ".join(results)
    
    return context
```
